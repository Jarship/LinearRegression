### Perceptron
This is basic code for a Perceptron, which is based upon Neurons.

self.weights is equal to the number of inputs + 1. Can you guess why that is?

```python
class Perceptron(object):
    def __init__(self, no_of_inputs, threshold=100, learning_rate=0.01):
        self.threshold = threshold
        self.learning_rate = learning_rate
        self.weights = np.zeros(no_of_inputs + 1)
        
    def predict(self, inputs):
        # A perceptron has a threshold (Here, summation) that has to be 
        # overcome in order to propogate an impulse.
        summation = np.dot(inputs, self.weights[1:]) + self.weights[0]
        if summation > 1:
            return 1
        else:
            return 0
    
    def train(self, training_inputs, labels):
        for _ in range(self.threshold):
            for inputs, label in zip(training_inputs, labels):
                prediction = self.predict(inputs)
                self.weights[1:] += self.learning_rate * (label - prediction) * inputs
                self.weights[0] += self.learning_rate * (label - prediction)
```

#### Training

The training step is an important step.
Here, the perceptron will loop a given number of times (threshold).
Each time, it will run through each set of inputs and outputs.
It will make a prediction of each set, based upon it's weights
(predict function)
Then each weight will be updated based upon 
    1. The error (label - prediction)
    2. The inputs, for all weights but the bias. (Any guesses why not?)
    3. The learning rate, which is a low number. (Any guesses why?)

When you have a given function,
represented here in point slope form,

##### y = m x + b

and you are adjusting the values that are not independent or dependent variables, then you have to make adjustments based upon predictions and errors from that prediction.

Since we know what y actually is, then we can subtract them.
Predict based upon our weights.

##### ŷ = mx + b

Then calculate the error:
E = y - ŷ

If the calculated value, ŷ, is larger than y, then E will be negative. We will want to shift the calculation down, to reduce the error.

If the calculated value is smaller than y, then E will be positive. We will want to shift the calculation up, to reduce the error.

This means that the weights should be shifted based upon some multiple of the Error.

How do we know how much to scale the Error by for each weight?

We can turn to Derivatives from Calculus.

A Derivative is the slope of a function.

[Derivative as a concept. Khan Academy](https://www.youtube.com/embed/N2PpRnFqnqY) 


With f(x) = m x + b

The derivative of f(x) in terms of m is x.
The derivative of f(x) in terms of b is 1.

The change to the bias is the learning rate * E * 1
The change to the weights(m) is the learning rate * E * x

```python
# A vector of numbers, from -25 to 24
x = np.linspace(-25,24)
# A vector of zeros of similar size to X
y = np.zeros_like(x)
# Setting all values from index 26 and up to 1
y[26:] = 1
```

```python
# As long as x is 0 or less, y is 0
# As long as x is 1 or more, y is 1
for v in range(len(x)):
    print('X value ' + str(v) + ' is ' + str(x[v]) +'. Y value ' + str(v) + ' is ' + str(y[v]))
```

```python
# Create a perceptron
p = Perceptron(1)
# Train it on x and y
p.train(x,y)
```

Using point slope form as an example

##### y = mx + b

We use two values to move from input of X to output of Y.

The bias, or b is just added to the final value.
The slope, or m, is a value that we use to scale the input into the range of the output. These numbers are represented in the predict function.

self.weight[0] is the bias

self.weight[1:] is the slope

```python
# A prediction of any number, 0 or lower, should result in a 0
p.predict(0)
```

```python
# A prediction of any number, 1 or higher, should result in a 1
p.predict(1)
```

### And Gate

```python
# recreating the truth table
training_inputs = []
training_inputs.append(np.array([0, 0]))
training_inputs.append(np.array([0, 1]))
training_inputs.append(np.array([1, 0]))
training_inputs.append(np.array([1, 1]))

labels = np.array([0, 0, 0, 1])
```

```python
# Creating a perceptron with two inputs and training it on the table
perceptron = Perceptron(2)
perceptron.train(training_inputs, labels)
```

Going back to point slope form,
##### y = mx + b

Now our input is a vector of length 2.

So there are three weights.

y = [0, 0] * x + [0]

with self.weights[0] as the bias

and self.weights[1:] as the slope.

```python
inputs = np.array([1, 1])
perceptron.predict(inputs)
```

```python
inputs = np.array([0, 1])
perceptron.predict(inputs) 
#=> 0
```

```python
perceptron.weights
```

Here, we can see that the function trained to have

##### y = [0.34, 0.34] * x + [0.34]

```python
inputs = np.reshape(training_inputs,(4,2))
```

```python
# With a scatter plot, we can see with colors whether the result is 
# 1 or 0.
plt.scatter(inputs[:,0],inputs[:,1], c=labels)
plt.axis([-0.1, 1.1, -0.1, 1.1])
plt.show()
```

```python
# Now, let's visualize how this function was an approximation
# Let's create a grid from 0 to 1 in two dimensions, with a step size of .005
xy = np.mgrid[0:1:0.005, 0:1:0.005].reshape(2,-1).T
# Then let's create a value that tracks the result of the function
ran = np.zeros(len(xy))
# For each combination
for num in range(len(ran)):
    # Predict that combinations outcome
    ran[num] = perceptron.predict(xy[num])
```

```python
plt.scatter(xy[:,0],xy[:,1], c = ran)
plt.axis([-0.1, 1.1, -0.1, 1.1])
plt.show()
```

```python
# The Data we are using is the Parts per million level of carbon dioxide
# in the atmosphere, from 1985 to 2018
df = pd.read_csv('./Datahub/co2-mm-mlo_csv.csv')
```

```python
df.head(15)
```

```python
# Pull the data we want from the data frame
dates,ppm = np.array(df['Date']),np.array(df['Interpolated'])
```

```python
# Plot it on a scatter plot
plt.scatter(dates,ppm)
# Nice linear correlation
plt.show()
```

```python
# A function for plotting the ppm by date
def plot_carbonppm(time, carbon, weights = None):
    plt.scatter(time, carbon)
    plt.rcParams['figure.figsize'] = [10, 5]
    plt.tick_params(axis='x', which='major', labelsize=12)
    if weights is not None:
        plt.plot(time, weights[0] + weights[1] * time, 'r-')
    plt.show()
```

```python
plot_carbonppm(dates, ppm)
```

### Neural Network

A Neural Network is similar to a perceptron, but it is a continuous output, rather than a binary, threshold output.

In this case, we are going to show a simple linear activation.

```python
class NeuralNetwork(object):
    
    def __init__(self, no_of_inputs, learning_rate=0.001):
        self.learning_rate = learning_rate
        self.weights = np.zeros(no_of_inputs + 1)
        
    def predict(self, inputs):
        # No threshold here
        return np.dot(inputs, self.weights[1:]) + self.weights[0]
    
    def train(self, training_inputs, labels):
        i = 0
        for inputs, label in zip(training_inputs, labels):
            prediction = self.predict(inputs)
            error = (label - prediction)
            self.weights[1:] += self.learning_rate * error * inputs
            self.weights[0] += self.learning_rate * error
```

```python
# Changing dates to numbers
for x in range(len(dates)):
    # Casting as Decimal, months divided by 12
    dates[x] = float(dates[x][:4]) + float(dates[x][5:7])/12
```

```python
# scikit learn is a basic machine learning library
# MinMaxScaler will take my inputs and outputs and scale them to 
# a given range

# This will help the neural network to learn. 
# Specifically, the backpropogation
from sklearn.preprocessing import MinMaxScaler 
```

```python
# The given range is 0 to 1
scaler = MinMaxScaler(feature_range=(0,1))
# Reshape and scale the y
co2 = scaler.fit_transform(ppm.reshape(-1,1))
# Reshape and scale the x
year = scaler.fit_transform(dates.reshape(-1,1))
```

```python
# Rescaled has not lost the linear relationship
plot_carbonppm(year,co2)
```

```python
# Creating a Neural Network with 1 input
net = NeuralNetwork(1)
# The number of times we will loop
epochs = 50
for i in range(epochs):
    # train given the data
    net.train(year,co2)
    # every five iterations of training, plot the line of best fit
    if i % 5 == 0:
        plot_carbonppm(year, co2, net.weights)
```
